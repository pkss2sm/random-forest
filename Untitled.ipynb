```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
from imblearn.over_sampling import SMOTE

# ১. ডাটা লোড করা (উদাহরণ হিসেবে একটি ডামি ডাটা তৈরি করা হয়েছে)
# আপনার নিজের CSV থাকলে pd.read_csv('filename.csv') ব্যবহার করুন
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_features=10, n_clusters_per_class=1,
                           weights=[0.9, 0.1], flip_y=0, random_state=42)

# ২. ট্রেইন এবং টেস্ট সেটে ভাগ করা
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"SMOTE এর আগে পজিটিভ ক্লাসের সংখ্যা: {sum(y_train == 1)}")

# ৩. SMOTE প্রয়োগ করা (শুধুমাত্র ট্রেনিং ডাটাতে করতে হয়)
sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)

print(f"SMOTE এর পরে পজিটিভ ক্লাসের সংখ্যা: {sum(y_train_res == 1)}")

# ৪. Random Forest মডেল তৈরি এবং ট্রেইন করা
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_res, y_train_res)

# ৫. প্রেডিকশন এবং রেজাল্ট দেখা
y_pred = rf_model.predict(X_test)

print("\n--- Confusion Matrix ---")
print(confusion_matrix(y_test, y_pred))

print("\n--- Classification Report ---")
print(classification_report(y_test, y_pred))
print("\n-- Accuracy Score--")
print(f"Accuracy Score: {accuracy_score(y_test, y_pred) * 100:.2f}%")
```

```python

import pandas as pd
import matplotlib.pyplot as plt

# ১. মডেলের ফিচার সংখ্যা কত সেটা দেখে নিই
num_features_in_model =rf_model.n_features_in_
print(f"মডেলে মোট ফিচার আছে: {num_features_in_model} টি")

# ২. তোমার বর্তমান X এর কলামগুলোর নাম ঠিক করি
# যদি X এখন array হয়, তবে আমরা df_clean থেকে নাম নেব
feature_names = data_clean.drop('Target', axis=1).columns

# ৩. যদি সংখ্যায় অমিল থাকে (Encoding এর কারণে), তবে আমরা শুধু সিকোয়েন্স অনুযায়ী নাম দেব
if len(feature_names) != num_features_in_model:
    # অনেক সময় encoding এর পর কলাম বেড়ে যায়, সেক্ষেত্রে আমরা generic নাম দেব
    feature_names = [f"Feature {i}" for i in range(num_features_in_model)]

# ৪. এবার ডাটাফ্রেম তৈরি
importances = rf_model.feature_importances_
feature_importance_df = pd.DataFrame({
    'Feature': feature_names, 
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

print("\nTop 10 Important Features:")
print(feature_importance_df.head(10))

# ৫. গ্রাফ
plt.figure(figsize=(10, 6))
plt.barh(feature_importance_df['Feature'].head(10), feature_importance_df['Importance'].head(10))
plt.gca().invert_yaxis()
plt.show()
```
